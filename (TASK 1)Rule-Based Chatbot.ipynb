```python
import nltk
import re
import random

print(" WELCOME TO THE RULE - BASED CHATBOT ")

# Define rules and responses for the chat bot
chat_rules = {
    r"(hello|hi|hey)" : ["Hello! How can I assist you today?", "Hi there!", "Hey, how can I help?"],
    r"I am fine, thank you" : ["Great to hear that:) How can I help you"],
    r"what is artificial intelligence?" : ["""Artificial intelligence (AI) is the basis for mimicking human intelligence processes through the creation
    and application of algorithms built into a dynamic computing environment. Stated simply, AI is trying to make computers think and act like humans."""],
    r"thank you so much, that was helpful" : ["I am happy to help :) "],
    r"how are you" : ["I'm just a bot, but thanks for asking!", "I'm doing well, thanks!"],
    r"what is your name" : ["You can call me ChatBot :)"],
    r"goodbye|bye" : ["Goodbye! Have a great day and see you soon"],
}

def chat_bot(message):
    for pattern, responses in chat_rules.items():
        if re.search(pattern, message):
            res = ' '.join(random.choices(responses))
            return res

# Test the chat bot
while True:
    message = input("You: ")
    if message.lower() == "exit":
        break
    response = chat_bot(message)
    print("Bot:", response)
```
